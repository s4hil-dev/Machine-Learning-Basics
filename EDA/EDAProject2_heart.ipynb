```python
def plotting(var,num):
  plt.subplot(2,2,num)
  sns.histplot(df[var],kde=True)
  plt.title(var)
plotting('Age',1)
plotting('RestingBP',2)
plotting('Cholesterol',3)
plotting('MaxHR',4)
plt.tight_layout() # to remove overlapping
```

There is RestingBP value 0 which is invalid

```python
restingBP_mean = df.loc[df['RestingBP']!=0,'RestingBP'].mean()
restingBP_mean
```

```python
df['RestingBP']= df['RestingBP'].replace(0,restingBP_mean)
df['RestingBP'] = df['RestingBP'].round(2)
```

```python
df['Cholesterol'].value_counts()
```

0 Cholestrol is also invald

```python
ch_mean = df.loc[df['Cholesterol']!=0,'Cholesterol'].mean()
ch_mean
```

```python
df['Cholesterol'] = df['Cholesterol'].replace(0,ch_mean)
df['Cholesterol'] = df['Cholesterol'].round(2)
```

```python
def plotting(var,num):
  plt.subplot(2,2,num)
  sns.histplot(df[var],kde=True)
  plt.title(var)
plotting('Age',1)
plotting('RestingBP',2)
plotting('Cholesterol',3)
plotting('MaxHR',4)
plt.tight_layout() # to remove overlapping
```

# Data Preprocessing

```python
df_encode = pd.get_dummies(df,drop_first=True)
df_encode
```

```python
df_encode = df_encode.astype(int)
df_encode
```

```python
from sklearn.preprocessing import StandardScaler
numerical_cols =['Age', 'RestingBP', 'Cholesterol','MaxHR','Oldpeak']
scaler = StandardScaler()
df_encode[numerical_cols] = scaler.fit_transform(df_encode[numerical_cols])
df_encode.head()
```
